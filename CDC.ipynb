```python
import pandas as pd
```

```python
data = pd.read_csv("TeleUse.csv") 
```

```python
indicator_race = data.loc[data["Subgroup"] !="Total"], data[["Indicator", "Group"]]   
```

```python
OffersTele = data[(data.Group == 'Race/Hispanic origin') & (data.Group != 'Total') & (data.Response !="Total") & (data.Indicator == "Provider offers telemedicine")]
OffersTeleBefore = data[(data.Group == 'Race/Hispanic origin')  & (data.Group != 'Total') & (data.Response !="Total") & (data.Indicator == "Provider offered telemedicine prior to pandemic")]
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(8,6))
plt.bar(OffersTele['Response'],  OffersTele['Percent'])
plt.xlabel('Respnse')
plt.ylabel('Percent') 
plt.title('RANDS - Provider Offers Telemedicine')
plt.xticks(color='black', rotation=45, horizontalalignment='right')
```

```python
OffersTeleBefore = data[(data.Group == 'Race/Hispanic origin')  & (data.Group != 'Total') & (data.Response !="Total") & (data.Indicator == "Provider offered telemedicine prior to pandemic")]
fig, ax = plt.subplots(figsize=(8,6))
plt.bar(OffersTeleBefore['Response'],  OffersTeleBefore['Percent'])
plt.xlabel('Respnse')
plt.ylabel('Percent') 
plt.title('RANDS - Telemedicine before pandemic All Races')
plt.xticks(color='black', rotation=45, horizontalalignment='right')
      
    # visualizing the plot
plt.show()
```
